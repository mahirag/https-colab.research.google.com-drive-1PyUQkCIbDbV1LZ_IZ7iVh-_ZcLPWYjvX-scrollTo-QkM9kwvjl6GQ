```python
import pandas as pd
```

```python
filename = '/content/drive/My Drive/titanic (1).csv'
df = pd.read_csv(filename)
df
```

```python
df.head(25)
```

```python
df_mean = df["Survived"].mean()
print(df_mean)
```

```python
table = pd.crosstab(df['Survived'], df['Sex'])
print(table)
```

```python
df_mean = df[df.Fare <= 10]['Fare'].mean()
print(df_mean)
```

```python
dead_mean = df[df['Survived'] == 0]['Age'].mean()
print(dead_mean)
```

```python
dead_mean = df[df['Survived'] == 1]['Age'].mean()
print(dead_mean)
```

```python
dead_male_mean = df[(df['Survived'] == 1) & (df['Sex'] == "male")]['Age'].mean()

print(dead_male_mean)
```
